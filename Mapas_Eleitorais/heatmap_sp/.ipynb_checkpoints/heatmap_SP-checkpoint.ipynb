# Heatmap São Paulo

```python
import pandas as pd
import numpy as np
import folium
from folium import plugins
import branca.colormap
from collections import defaultdict
from folium.plugins import HeatMap
```

```python
base_sp = pd.read_csv('base_heatmap_sp.csv')
```

```python
cidades = pd.read_csv('https://raw.githubusercontent.com/kelvins/Municipios-Brasileiros/main/csv/municipios.csv')
```

```python
# visualizando o data frame
base_sp.head()
```

```python
cidades.head()
```

```python
# Visualizando as informações do data frame
base_sp.info()
```

```python
# Mudando o tipo da coluna para numero 
base_sp['repr_pret_eleitos'] = base_sp['repr_pret_eleitos'].apply(lambda x: float(x.replace(".","").replace(",",".")))
```

```python
base_sp['repr_negr_eleitos'] = base_sp['repr_negr_eleitos'].apply(lambda x: float(x.replace(".","").replace(",",".")))
```

```python
base_sp['inves_negr100'] = base_sp['inves_negr100'].apply(lambda x: float(x.replace(".","").replace(",",".")))
```

```python
base_sp.drop(['municipios'], axis=1, inplace = True)
```

```python
base_sp.rename(columns={'Unnamed: 1': 'municipios'}, inplace=True)
```

```python
cities_sp = cidades.loc[cidades['codigo_uf'] == 35]
```

```python
cities_sp.rename(columns={'nome': 'municipios'}, inplace=True)
```

```python
cities_sp.info()
```

```python
cities_sp
```

```python
geoloc_base_sp = pd.merge(base_sp, cities_sp, on='municipios')
```

```python
geoloc_base_sp.head()
```

```python
geoloc_base_sp.info()
```

```python
geoloc_base_sp.drop([ 'capital', 'codigo_uf'],  axis=1, inplace= True)
```

```python
geoloc_base_sp.to_csv('geoloc_base_sp.csv')
```

```python
geoloc_base_sp['repr_pret_eleitos'] = geoloc_base_sp['repr_pret_eleitos'].replace([244307.00], 244)
```

```python
geoloc_base_sp['repr_pret_eleitos'] = geoloc_base_sp['repr_pret_eleitos'].replace([228735.0], 229)
```

```python
geoloc_base_sp['inves_negr100'] = geoloc_base_sp['inves_negr100'].replace([29375.0], 294)
```

```python
geoloc_base_sp['inves_negr100'] = geoloc_base_sp['inves_negr100'].replace([11875.0], 119)
```

```python
geoloc_base_sp['inves_negr100'] = geoloc_base_sp['inves_negr100'].replace([56875.0], 569)
```

```python
geoloc_base_sp.to_csv('geoloc_base_sp.csv')
```

#### Heatmap Representatividade Pretos Eleitos

```python
coordenadas1 = geoloc_base_sp[['latitude', 'longitude', 'repr_pret_eleitos']]
```

```python
coordenadas1.info()
```

```python
geoloc_base_sp['repr_pret_eleitos'].describe()
```

```python
map_1 = folium.Map(
                width= '100%',
                heigth = '100%',
                location = [-23.5489, -46.6388],
                zoom_start = 8)
```

```python
steps=20
colormap = branca.colormap.linear.YlOrRd_09.scale(0, 1).to_step(steps)
colormap.caption = 'Representatividade Pretos Eleitos'
gradient_map=defaultdict(dict)
for i in range(steps):
    gradient_map[1/steps*i] = colormap.rgb_hex_str(1/steps*i)
colormap.add_to(map_1) 
```

```python
map_1 = map_1.add_child(plugins.HeatMap(coordenadas1,fill_opacity = 0.5, gradient = gradient_map, 
                                            show=True))
```

```python
for i in range(0, len(geoloc_base_sp)):
    folium.Circle(
        location = [geoloc_base_sp.iloc[i]['latitude'] , geoloc_base_sp.iloc[i]['longitude']],
        color = '#00FF69',
        fill = '#00A1B3',
        tooltip = '<li><bold> MUNICIPIO:' +str(geoloc_base_sp.iloc[i]['municipios']) + "</bold><li>"+
                  '<li><bold> Representatividade:' + str(geoloc_base_sp.iloc[i]['repr_pret_eleitos']) + "</bold><li>",
        radius = (geoloc_base_sp.iloc[i]['repr_pret_eleitos']**1.1)
    ).add_to(map_1)
```

```python
map_1
```

```python
map_1.save('repr_pretos_eleitos_sp.html')
```
