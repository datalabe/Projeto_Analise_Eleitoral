# Heatmap Bahia


```python
import pandas as pd
import numpy as np
import folium
from folium import plugins
import branca.colormap
from collections import defaultdict
from folium.plugins import HeatMap
```

```python
base_ba = pd.read_csv('base_heatmap_ba.csv')
```

```python
cidades = pd.read_csv('https://raw.githubusercontent.com/kelvins/Municipios-Brasileiros/main/csv/municipios.csv')
```

```python
# visualizando o data frame
base_ba.head()
```

```python
cidades.head()
```

```python
# Visualizando as informações do data frame
base_ba.info()
```

```python
base_ba.drop(['maisculo'], axis=1, inplace = True)
```

```python
base_ba['repr_negra100'] = base_ba['repr_negra100'].apply(lambda x: float(x.replace(".","").replace(",",".")))
```

```python
cities_ba = cidades.loc[cidades['codigo_uf'] == 29]
```

```python
cities_ba.rename(columns={'nome': 'municipios'}, inplace=True)
```

```python
cities_ba.info()
```

```python
cities_ba
```

```python
geoloc_base_ba = pd.merge(base_ba, cities_ba, on='municipios')
```

```python
geoloc_base_ba.head()
```

```python
geoloc_base_ba.info()
```

```python
geoloc_base_ba.drop([ 'capital', 'codigo_uf'],  axis=1, inplace= True)
```

```python
geoloc_base_ba.to_csv('geoloc_base_ba.csv')
```

#### Heatmap Representatividade Pretos Eleitos

```python
coordenadas1 = geoloc_base_ba[['latitude', 'longitude', 'repr_pret_eleitos']]
```

```python
coordenadas1.info()
```

```python
geoloc_base_ba['repr_pret_eleitos'].describe()
```

```python
map_1 = folium.Map(
                width= '100%',
                heigth = '100%',
                location = [-13.3945273,-46.479903],
                zoom_start = 8)
```

```python
steps=20
colormap = branca.colormap.linear.YlOrRd_09.scale(0, 1).to_step(steps)
colormap.caption = 'Representatividade Pretos Eleitos'
gradient_map=defaultdict(dict)
for i in range(steps):
    gradient_map[1/steps*i] = colormap.rgb_hex_str(1/steps*i)
colormap.add_to(map_1) 
```

```python
map_1 = map_1.add_child(plugins.HeatMap(coordenadas1,fill_opacity = 0.5, gradient = gradient_map, 
                                            show=True))
```

```python
for i in range(0, len(geoloc_base_ba)):
    folium.Circle(
        location = [geoloc_base_ba.iloc[i]['latitude'] , geoloc_base_ba.iloc[i]['longitude']],
        color = '#00FF69',
        fill = '#00A1B3',
        tooltip = '<li><bold> MUNICIPIO:' +str(geoloc_base_ba.iloc[i]['municipios']) + "</bold><li>"+
                  '<li><bold> Representatividade:' + str(geoloc_base_ba.iloc[i]['repr_pret_eleitos']) + "</bold><li>",
        radius = (geoloc_base_ba.iloc[i]['repr_pret_eleitos']**1.1)
    ).add_to(map_1)
```

```python
map_1
```

```python
map_1.save('repr_pretos_eleitos_ba.html')
```
